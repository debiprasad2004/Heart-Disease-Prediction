```python
#Load create of the data frame of the set mentioned
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Heart_disease_cleveland_new.csv')
df.head(10)
```

```python
df.shape
```

```python
df.isnull().any()
```

```python
y = df.target
x = df.drop(['target'], axis=1)
print(x.shape)
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input
model=Sequential([
    Input(shape=(13,)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score
y_pred = model.predict(x_test)
hist_mod = model.fit(x_train, y_train, epochs=200, batch_size=32, validation_data=(x_test, y_test), verbose=1)
```

```python
acc_mod = model.evaluate(x_test, y_test, verbose=1)
print('Test Accuracy:', acc_mod[1])
```

```python
Patient1= ([55, 1, 2, 140, 220, 0, 1, 150, 0, 1.5, 2, 0, 3])
patient1 = np.array(Patient1)
patient1_scaled = scaler.transform(patient1.reshape(1, -1))
prediction = model.predict(patient1_scaled)
print("Prediction:", prediction)
if prediction > 0.8:
  print("The patient has heart disease")
else:
  print("The patient does not have heart disease")
```
